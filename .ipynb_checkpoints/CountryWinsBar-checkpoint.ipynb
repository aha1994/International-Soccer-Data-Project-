```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import geopy
import requests
import json
import citipy
```

```python
# Read in csv
countryWins = pd.read_csv('cleanedSoccerDF.csv')
countryWins.head()
```

```python
countryWinsDF = pd.DataFrame(countryWins['Result'].value_counts())
countryWinsDF = countryWinsDF.drop('Draw')
countryWinsDF
```

```python
# Bar graph for country wins
x_axis = countryWinsDF.index
x_ticks = np.arange(0, len(x_axis)) * 1.5
y = countryWinsDF['Result']

plt.figure(figsize=(80,10))
plt.bar(x_ticks, y, align='edge')
plt.xticks(x_ticks, x_axis, rotation='vertical')
plt.xlabel('Country')
plt.ylabel('Number of Wins')
plt.title('Number of International Wins By Country')
plt.xlim(-1, 340)

plt.savefig('Country_Wins.png')
plt.show()
```

```python
# Top 25 wins by countries in matplotlib bar graph

top20DF = countryWinsDF.iloc[0:25,:]
top20DF

x_axis = top20DF.index
x_ticks = np.arange(0, len(x_axis)) * 1.5
y = top20DF['Result']

plt.figure(figsize=(10,10))
plt.bar(x_ticks, y, align='edge')
plt.xticks(x_ticks, x_axis, rotation='vertical')
plt.xlabel('Country')
plt.ylabel('Number of Wins')
plt.title('Number of International Wins By Country Top 25')
plt.xlim(-1, 37.5)

plt.savefig('Top20_Country_Wins.png')
plt.show()
```
