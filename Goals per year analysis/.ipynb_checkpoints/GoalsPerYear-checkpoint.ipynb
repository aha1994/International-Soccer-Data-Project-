```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import geopy
import requests
import json
import citipy
```

```python
# Read in csv
df1 = pd.read_csv('Updated_cleaned_DF.csv')
df1.head()
# Group by date
group_year = df1.groupby('date')
totalGoalDF = pd.DataFrame(group_year['total goals'].mean())
x_axis = totalGoalDF.index
y_val = totalGoalDF['total goals']

#Display scatter plot with matplotlib
plt.scatter(x_axis, y_val)
plt.xlabel('Year')
plt.ylabel('Average Total Goals Per Match')
plt.title('Average Total Goals Scored Over Time')
plt.savefig('Average Total Goals Scored Over Time.png')
plt.show()
```
